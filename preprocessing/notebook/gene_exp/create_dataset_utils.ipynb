```python
import numpy as np
import pandas as pd
import os
```

```python
input_dir = '/csc/epitkane/projects/PCAWG_geneexp/' 
```

```python
export_dir = '/csc/epitkane/projects/PCAWG_geneexp/'
```

```python
pcawg_gx = pd.read_csv(input_dir + 'PCAWG_gene_expression.tsv',sep='\t',index_col=0)
```

```python
pcawg_gx
```

```python
#create classinfo_(name).csv

count_histology = pcawg_gx['histology'].value_counts()
```

```python
pd_data = pd.DataFrame(count_histology)
```

```python
pd_data = pd_data.reset_index()
```

```python
pd_data.columns = ['histology','n_samples']
```

```python
pd_data
```

```python
pcawg_all_dir = '/csc/epitkane/projects/PCAWG20191001/data/modified_data/raw/indelmeisubsvperclass/'
```

```python
pcawg_all = os.listdir(pcawg_all_dir)
```

```python
pcawg_all
```

```python
pd_pcawg_info = pd.DataFrame()
for i in pcawg_all: 
    num_samples = len(os.listdir(pcawg_all_dir + '/' + i))
    row = [i, num_samples]
    
    pd_data = pd.DataFrame(row).T
    pd_pcawg_info = pd_pcawg_info.append(pd_data)
```

```python
pd_pcawg_info.columns = ['histologywgs','n_samples']
```

```python
pd_pcawg_info = pd_pcawg_info.reset_index(drop=True)
```

```python
pd_pcawg_info
```

```python
wgs_gx_info = pd_pcawg_info.merge(pd_data,left_on='histologywgs',right_on='histology',how='inner')
```

```python
wgs_gx_info.columns = ['histology_wgs','n_samples_wgs','histology_gx','n_samples_gx']
```

```python
wgs_gx_info = wgs_gx_info.sort_values(by=['histology_wgs'])
```

```python
#select only gx with more than 10 samples

wgs_gx_info = wgs_gx_info.loc[wgs_gx_info['n_samples_gx']>10]
```

```python
wgs_gx_info
```

```python
wgs_gx_info = wgs_gx_info.reset_index(drop=True)
```

```python
wgs_gx_info = wgs_gx_info.reset_index()
```

```python
wgs_gx_info = wgs_gx_info.rename(columns={"index": "class_index"})
```

```python
wgs_gx_info.to_csv(export_dir + 'wgs_gx_info.tsv',sep='\t')
```

```python
wgs_gx_info
```

```python
#create classinfo_(name).csv

selected = wgs_gx_info[['histology_gx','class_index','n_samples_gx']]
```

```python
selected.columns = ['class_name','class_index','n_samples']
```

```python
selected
```

```python
selected.to_csv('../../../dataset_utils/classinfo_wgsgx.csv')
```

```python
selected_class = selected['class_name'].to_list()
```

```python
selected_class
```

```python
selected_gx = pcawg_gx.loc[pcawg_gx['histology'].isin(selected_class)]
```

```python
selected_gx.to_csv(export_dir + 'selected_gx.csv')
```

```python
allsamples_histology = selected_gx[['samples','histology']]
```

```python
allsamples_histology.columns = ['samples','nm_class']
```

```python
get_10slices = []
startslice = 0

for i in range(0,len(allsamples_histology)):
    startslice = startslice+1
    if startslice > 10:
        startslice = 1
        get_10slices.append(startslice)
    else:
        get_10slices.append(startslice)
allsamples_histology['slices'] = get_10slices
```

```python
#create_train_val_test split,
trainAll = pd.DataFrame()
valAll = pd.DataFrame()

for valfold in range(1,11):
    val = allsamples_histology.loc[allsamples_histology['slices']==valfold]
    train = allsamples_histology.loc[allsamples_histology['slices']!=valfold]
    
    train['fold'] = valfold
    val['fold'] = valfold
    
    trainAll = trainAll.append(train)
    valAll = valAll.append(val)
trainAll.to_csv('../../../dataset_utils/' + 'wgsgx_train.csv')
valAll.to_csv('../../../dataset_utils/' + 'wgsgx_val.csv')
```
