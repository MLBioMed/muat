```python
import numpy as np
import pandas as pd
import os
```

```python
root = '/csc/epitkane/data/PCAWG/transcriptome/gene_expression/'
```

```python
export_dir = '/csc/epitkane/projects/PCAWG_geneexp/' 
```

```python
filename = 'tophat_star_fpkm_uq.v2_aliquot_gl.tsv.gz'
```

```python
pd_data = pd.read_csv(root + filename,sep='\t')
```

```python
pd_data = pd_data.T
```

```python
new_header = pd_data.iloc[0] #grab the first row for the header
pd_data = pd_data[1:] #take the data less the header row
pd_data.columns = new_header
```

```python
pd_data = pd_data.reset_index()
```

```python
pd_data = pd_data.rename(columns={"index": "sample"})
```

```python
pd_data
```

```python
PCAWG_metadata = '/csc/epitkane/projects/PCAWG20191001/data/pcawg2653.data.matrix.Feb2017.v1.tsv'
```

```python
msi_meta = '/csc/epitkane/data/PCAWG2019/release_may2016.v1.4.tsv'
```

```python
pd_meta = pd.read_csv(msi_meta,sep='\t')
```

```python
pd_meta = pd_meta[['donor_unique_id','submitter_donor_id','icgc_donor_id','normal_wgs_submitter_specimen_id','tumor_rna_seq_icgc_sample_id','tumor_rna_seq_aliquot_id','tumor_wgs_aliquot_id']]
```

```python
pd_PCAWG_meta = pd.read_csv(PCAWG_metadata,sep='\t')
```

```python
pd_PCAWG_meta
```

```python
pd_PCAWG_meta = pd_PCAWG_meta[['tumor_wgs_aliquot_id','histology']]
```

```python
pd_PCAWG_meta
```

```python
pd_meta = pd_meta.merge(pd_data,left_on='tumor_rna_seq_aliquot_id',right_on='sample', how='inner')
```

```python
pd_meta.to_csv(export_dir + 'metawith_geneexp.tsv',sep='\t')
```

```python
gene_histo = pd_PCAWG_meta.merge(pd_meta,left_on='tumor_wgs_aliquot_id',right_on='tumor_wgs_aliquot_id',how='inner')
```

```python
gene_histo.to_csv(export_dir + 'gene_histology.tsv',sep='\t')
```

```python
only_geneexp = gene_histo.iloc[:,9:]
```

```python
only_geneexp
```

```python
only_geneexp['samples'] = gene_histo['tumor_wgs_aliquot_id']
```

```python
only_geneexp['histology'] = gene_histo['histology']
```

```python
only_geneexp.to_csv(export_dir + 'PCAWG_gene_expression.tsv',sep='\t')
```
